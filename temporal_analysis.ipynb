```python
import pandas as pd

# Read the JSON file
df = pd.read_json("/home/hampusnasstrom/Downloads/workflows-1000-1769174146067.json")

# Display the dataframe
print(df)
```

```python
# Expand the workflows column into separate columns
df_expanded = pd.json_normalize(df['workflows'])

print(df_expanded.head())
print(f"\nShape: {df_expanded.shape}")
print(f"\nColumns: {df_expanded.columns.tolist()}")
```

```python
# Filter for ProcessEntryWorkflow and calculate time metrics
process_entry = df_expanded[df_expanded['name'] == 'ProcessEntryWorkflow'].copy()
print(f"ProcessEntryWorkflow rows: {len(process_entry)}")

# Convert timestamps to datetime
process_entry['startTime'] = pd.to_datetime(process_entry['startTime'])
process_entry['endTime'] = pd.to_datetime(process_entry['endTime'])

# Calculate duration in milliseconds
process_entry['duration_ms'] = (process_entry['endTime'] - process_entry['startTime']).dt.total_seconds() * 1000

print("\nFirst 5 ProcessEntryWorkflow entries with calculated duration:")
print(process_entry[['id', 'startTime', 'endTime', 'duration_ms']].head())

print(f"\n\nDuration Statistics for ProcessEntryWorkflow (milliseconds):")
print(process_entry['duration_ms'].describe())

print(f"\nMin duration: {process_entry['duration_ms'].min():.2f} ms")
print(f"Max duration: {process_entry['duration_ms'].max():.2f} ms")
print(f"Median duration: {process_entry['duration_ms'].median():.2f} ms")
```
